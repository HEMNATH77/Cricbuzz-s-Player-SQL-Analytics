```python
import requests
import mysql.connector

# --- API Headers ---
headers = {
    "x-rapidapi-key": "bb8592df63msh6e0b1fcf0b4efe6p19071ajsn35e95eba42a0",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# --- Match IDs ---
match_ids = [
    122517, 100283, 100290, 130019, 82462, 116981, 116990, 
    76465, 105804, 87626, 87878, 116972, 69862, 43061, 
    76535, 100312, 60042, 105770,
    100328, 77894, 
    91778, 91787, 105780, 105770, 113864
    
]

count = 0
for mid in match_ids:
    try:
        url_scorecard = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{mid}/hscard"
        scorecard_data = requests.get(url_scorecard, headers=headers).json()

        if not scorecard_data:
            continue

        # --- Player name → ID mapping ---
        player_name_to_id = {}
        score_cards = scorecard_data.get("scorecard", [])

        for innings in score_cards:
            for player_info in innings.get("batsman", []) + innings.get("bowler", []):
                name = player_info.get("name")
                if name:
                    player_name_to_id[name.strip()] = player_info.get("id")

        # --- Extract fielding stats ---
        fielding_stats = {}
        for innings in score_cards:
            batsmen_list = innings.get("batsman", [])
            for bat_info in batsmen_list:
                out_desc = bat_info.get("outdec", "").strip()

                # --- Catch detection ---
                if out_desc.startswith("c "):
                    # Example: "c Kohli b Anderson"
                    parts = out_desc.split(" b ")
                    if len(parts) >= 2:
                        fielder = parts[0].replace("c", "").strip()
                        if fielder:
                            if fielder not in fielding_stats:
                                fielding_stats[fielder] = {"catches": 0, "stumpings": 0}
                            fielding_stats[fielder]["catches"] += 1

                # --- Stumping detection ---
                elif out_desc.startswith("st "):
                    # Example: "st Dhoni b Murali"
                    parts = out_desc.split(" b ")
                    if len(parts) >= 2:
                        keeper = parts[0].replace("st", "").strip()
                        if keeper:
                            if keeper not in fielding_stats:
                                fielding_stats[keeper] = {"catches": 0, "stumpings": 0}
                            fielding_stats[keeper]["stumpings"] += 1

        # --- Insert into DB ---
        for player_name, stats in fielding_stats.items():
            player_id = player_name_to_id.get(player_name, None)
            if player_id and (stats["catches"] > 0 or stats["stumpings"] > 0):
                cursor.execute("""
                    INSERT INTO fielding (
                        match_id, player_id, player_name, catches, stumpings
                    ) VALUES (%s,%s,%s,%s,%s)
                    ON DUPLICATE KEY UPDATE
                        catches=VALUES(catches),
                        stumpings=VALUES(stumpings)
                """, (mid, player_id, player_name, stats["catches"], stats["stumpings"]))
                count += 1

    except Exception as e:
        print(f"Failed to fetch fielding data for match {mid}: {e}")

conn.commit()
print(f"Inserted/Updated {count} fielding records successfully.")
```
